```python
import tensorflow as tf
```

```python
daily_sales_numbers = [21, 22, -108, 31, -1, 32, 34,31] #in usdollar

#creating a tensorflow - dataset 
tf_dataset = tf.data.Dataset.from_tensor_slices(daily_sales_numbers)
tf_dataset
```

```python
#this wud be in the tf format , so basically we need to convert it into numpy 
for sales in tf_dataset:
    print(sales)
```

```python
#1st approach
for sales in tf_dataset.as_numpy_iterator():
    print(sales)
```

```python
#2nd approach :- .take(3) function  selecting first 3 elts.
for sales in tf_dataset.take(3):
    print(sales.numpy())
```

```python
#as sale cant be in negative , so we need to filter it 
```

```python
#.filter()
tf_dataset = tf_dataset.filter(lambda x: x>0)
for sales in tf_dataset.as_numpy_iterator():
    print(sales)
```

```python
tf_dataset = tf_dataset.map(lambda x: x*72)
for sales in tf_dataset.as_numpy_iterator():
    print(sales)
```

```python
tf_dataset = tf_dataset.shuffle(2)
for sales in tf_dataset.as_numpy_iterator():
    print(sales)
```

```python
#Doing batching !
#buffer_size =2 means batch of 2
for sales_batch in tf_dataset.batch(2):
    print(sales_batch.numpy())
```

```python
#this get_label will select from the string the word cat by splitting it with \\ 
def get_label(file_path):
    import os
    parts = tf.strings.split(file_path, os.path.sep)
    return parts[-2]
```

```python
get_label("images\\dog\\20 Reasons Why Cats Make the Best Pets....jpg")
```

```python
#the use of this get_label is to have a y , which is the target 
```

```python
def process_image(file_path):
    label = get_label(file_path)
    img = tf.io.read_file(file_path) # load the raw data from the file as a string
    img = tf.image.decode_jpeg(img)
    img = tf.image.resize(img, [128, 128])
    return img, label
```

```python
img, label = process_image(get_example())
img.numpy()[:2]
```

```python
train_ds = train_ds.map(process_image)
test_ds = test_ds.map(process_image)
```

```python
for image, label in train_ds.take(1):
    print("****",image)
    print("****",label)
```

```python
def scale(image, label):
    return image/255, label
```

```python
train_ds = train_ds.map(scale)
```
